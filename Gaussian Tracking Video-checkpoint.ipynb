```python
#Defining the functions used to track the particle
#2D Gaussian
def gaussian_2d(xy, amplitude, xo, yo, sigma_x, sigma_y, offset):
    x, y = xy
    g = offset + amplitude * np.exp(-(((x - xo)**2) / (2 * sigma_x**2) + ((y - yo)**2) / (2 * sigma_y**2)))
    
    # Flattening to a 1D arra6
    return g.ravel()

def track_particles(image, limit, expected_radius):
    # 1. ADJUST DETECTING SIZE: blob_log looks for particles based on 'sigma'
    # In LoG detection, the relationship is roughly: radius = sqrt(2) * sigma
    # So if you expect a particle radius of 8 pixels, max_sigma should be ~6.
    blobs = blob_log(image, min_sigma=2, max_sigma=expected_radius, num_sigma=5, threshold=limit)
    
    subpixel_centers = []
    
    for blob in blobs:
        y_int, x_int, radius = blob
        x_int, y_int = int(x_int), int(y_int)
        
        # 2. ADJUST CROPPING WINDOW: The box must be larger than the particle.
        # You need the whole particle PLUS a border of dark background pixels
        # so the curve_fit can accurately calculate the background offset.
        box_size = int(expected_radius * 1.5) 
        
        if (y_int - box_size < 0 or y_int + box_size >= image.shape[0] or 
            x_int - box_size < 0 or x_int + box_size >= image.shape[1]):
            continue
            
        roi = image[y_int - box_size : y_int + box_size, x_int - box_size : x_int + box_size]
        
        x = np.arange(x_int - box_size, x_int + box_size)
        y = np.arange(y_int - box_size, y_int + box_size)
        x_grid, y_grid = np.meshgrid(x, y)
        
        # 3. ADJUST INITIAL GUESS: Give the optimizer a realistic starting size.
        # Instead of '1.0, 1.0', we start the guess at half the expected radius.
        guess_sigma = expected_radius / 2.0
        initial_guess = (roi.max() - roi.min(), x_int, y_int, guess_sigma, guess_sigma, roi.min())
        
        try:
            popt, _ = curve_fit(
                gaussian_2d, 
                (x_grid, y_grid), 
                roi.ravel(), 
                p0=initial_guess,
                bounds=([0, x_int-box_size, y_int-box_size, 0.1, 0.1, 0], 
                        [np.inf, x_int+box_size, y_int+box_size, box_size, box_size, np.inf])
            )
            
            subpixel_centers.append((popt[1], popt[2]))
            
        except RuntimeError:
            continue
            
    return subpixel_centers
```

```python

# --- Assumes your updated track_particles function is defined above this ---

# 1. Setup Paths and Parameters
output_path = r'C:\Users\classes\Documents\260713_122322_659\tracked_output.mp4'

limit = 0.005             # Your background noise threshold
expected_radius = 8     # Your expected particle radius in pixels
fps = 15                # Frames per second for the output video

# 2. Open the TIFF file
height, width = frames[0].shape

# 3. Initialize the OpenCV Video Writer
# 'mp4v' is a widely compatible codec for MP4 files
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
video_writer = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

print(f"Starting tracking on {len(frames)} frames...")

center_list = []
uplimit = 1
downlimit = 1
adjust_exp = 0

# 4. Process Each Frame
for i, frame in enumerate(frames):
    # Pass the raw, unmodified frame to your math function
    adjusting = True
    crossover = False
    upprev = False
    downprev = False
    
    while adjusting == True:
        centers = track_particles(frame, limit, expected_radius)
        if len(centers) > 1:
            if downprev:
                adjust_exp += 1
            limit += 0.001*uplimit/(2**adjust_exp)
            uplimit += 1
            downlimit = 1
            upprev = True
            downprev = False
            print("Adjusting limit up. Limit = ", limit, ", uplimit = ", uplimit)
        elif not centers:
            if upprev:
                adjust_exp += 1
            limit -= 0.001*downlimit/(2**adjust_exp)
            downlimit += 1
            uplimit = 1
            downprev = True
            upprev = False
            print("Adjusting limit down. Limit = ", limit, ", downlimit = ", downlimit)
        else:
            adjusting = False
        if limit <= 0:
            limit = 0.0001
    if len(centers) > 0:
        x, y = centers[0]
        center_list.append([x, y])
    else:
        center_list.append(np.nan, np.nan)
    # --- Video Preparation Step ---
    # Scientific cameras often output 12-bit or 16-bit images. Standard video 
    # players only understand 8-bit images (pixel values from 0 to 255).
    # We must normalize the frame for visual export.
    frame_8bit = cv2.normalize(frame, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    
    # Convert the monochrome frame to BGR (color) so we can draw colored markers on it
    display_frame = cv2.cvtColor(frame_8bit, cv2.COLOR_GRAY2BGR)
    
    # 5. Annotate the Frame
    for (x, y) in centers:
        # OpenCV drawing functions require integer pixel coordinates
        x_int, y_int = int(round(x)), int(round(y))
        
        # Draw a red crosshair (BGR format: Blue=0, Green=0, Red=255)
        cv2.drawMarker(display_frame, (x_int, y_int), color=(0, 0, 255), 
                       markerType=cv2.MARKER_CROSS, markerSize=15, thickness=2)
        
    # 6. Add to Video
    video_writer.write(display_frame)
    
    # Simple progress tracker
    if (i + 1) % 200 == 0:
        print(f"Processed {i + 1} / {len(frames)} frames")


# 7. Finalize and Save
video_writer.release()
print(f"Tracking complete! Video saved to: {output_path}")
```

```python
center_list = np.array(center_list)

def calculate_msd(trajectory, max_lag):
    msd = []
    
    # Loop over all requested lag times (tau)
    for lag in range(1, max_lag + 1):
        
        # This one-liner finds the difference between every pair of points 
        # separated by 'lag' frames simultaneously.
        displacements = trajectory[lag:] - trajectory[:-lag]
        
        # Calculate the squared distance (dx^2 + dy^2) for each pair
        sq_dist = np.sum(displacements**2, axis=1)
        
        # np.nanmean ignores any frames where the particle was lost (NaNs)
        msd.append(np.nanmean(sq_dist))
        
    return np.array(msd)

# Calculate MSD for up to 50 frames of lag
max_lag = round(len(center_list)/2)

msd_pixels = calculate_msd(center_list, max_lag)

msd_mum = msd_pixels * 0.0907**2

lag_range = np.arange(1, max_lag+1)

plt.plot(lag_range, msd_mum)

plt.xlabel("Tau")
plt.ylabel("MSD / µm^2")
plt.show()

```
